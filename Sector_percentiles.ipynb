## Combine data

```python
nasdaq = pd.read_csv("tickers_data/nasdaq_stocks.csv")
nyse = pd.read_csv("tickers_data/nyse_stocks.csv")
amex = pd.read_csv("tickers_data/amex_stocks.csv")

exchange = np.array([], dtype="float64")
for i in range(len(nasdaq.index)):
    exchange = np.append(exchange, "nasdaq")
for i in range(len(nyse.index)):
    exchange = np.append(exchange, "nyse")
for i in range(len(amex.index)):
    exchange = np.append(exchange, "amex")
exchange

data = pd.concat([nasdaq, nyse, amex], axis=0)
len(data.index)
data["Exchange"] = exchange
data.to_csv("combined_tickers.csv")
```
